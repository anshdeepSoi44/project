```python
#distribution of weight and height
```

```python
plt.style.use('ggplot')
```

```python
data.Weight.plot(kind='hist',color='purple',edgecolor='black',figsize=(10,7))
plt.title('distbution of weight',size=24)
plt.xlabel('weight(pound)',size=18)
plt.ylabel('frequency',size=18);
```

```python
data.Height.plot(kind='hist',color='purple',edgecolor='black',figsize=(10,7))
plt.title('distbution of Height',size=24)
plt.xlabel('Height(inches)',size=18)
plt.ylabel('frequency',size=18);
```

```python
data[data['Gender']=='Male'].Height.plot(kind='hist',color='green',edgecolor='black',alpha=0.5,figsize=(10,7))
data[data['Gender']=='Female'].Height.plot(kind='hist',color='green',edgecolor='black',alpha=0.5,figsize=(10,7))
plt.legend(labels=['Males','Female'])
plt.title('distribution of Height',size=24)
plt.xlabel('Height(inches)',size=18)
plt.ylabel('frequency',size=18);
```

```python
data[data['Gender']=='Male'].Weight.plot(kind='hist',color='green',edgecolor='black',alpha=0.5,figsize=(10,7))
data[data['Gender']=='Female'].Weight.plot(kind='hist',color='green',edgecolor='black',alpha=0.5,figsize=(10,7))
plt.legend(labels=['Males','Female'])
plt.title('distribution of Weight',size=24)
plt.xlabel('Weight(pounds)',size=18)
plt.ylabel('frequency',size=18);
```

```python
#scatter plot of height and weight
```

```python
ax1=data[data['Gender']=='Male'].plot(kind='scatter',x='Height',y='Weight',color='blue',
alpha =0.5,figsize=(10,7))  
data[data['Gender']=='Female'].plot(kind='scatter',x='Height',y='Weight',color='magenta',
alpha =0.5,figsize=(10,7),ax=ax1)
plt.legend(labels=['Males','Females'])
plt.title('Relationship between Height and Weight',size=24)
plt.xlabel('Height(inches)',size=18)
plt.ylabel('Weight(pounds)',size=18);
```

```python
#scatter plot of 500 females
```

```python
sample_females=data[data['Gender']=='Female'].sample(50)
sample_females.plot(kind='scatter',x='Height',y='Weight',color='magenta',alpha=0.5,
figsize=(10,7))
plt.legend(labels=['Females'])
plt.title('Relationship between Height and Weight(sample of 50 females)',size=10)
plt.xlabel('Height(inches)',size=12)
plt.ylabel('Weight(pounds)',size=12);        
```
